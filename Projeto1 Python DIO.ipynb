```python
#Transformando coluna de object em data
df["Data Do Pedido"] = pd.to_datetime(df["Data Do Pedido"])
df["Data de Envio"] = pd.to_datetime(df["Data de Envio"])
```

```python
df.dtypes
```

```python
#Criando uma coluna com total de dias para enviar o produto
df["Tempo De Envio"] = df["Data de Envio"] - df["Data Do Pedido"]
```

```python
df.head()
```

```python
#Extraindo apenas os dias
df["Tempo De Envio"] = (df["Data de Envio"] - df["Data Do Pedido"]).dt.days
```

```python
df.head()
```

```python
#Verificando o tipo da coluna Tempo_envio
df['Tempo De Envio'].dtype
```

```python
#Verificando se temos dados faltantes
df.isnull().sum()
```

```python
#Vamos Agrupar por ano e marca
df.groupby([df["Data Do Pedido"].dt.year, "Categoria Do Produto"])["Lucro"].sum()
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Do Pedido"].dt.year, "Categoria Do Produto"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos vendido por categoria?
df.groupby("Categoria Do Produto")["Quantidade Pedida"].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de produtos vendidos
df.groupby("Categoria Do Produto")["Quantidade Pedida"].sum().sort_values(ascending=True).plot.barh(title="Total de Produtos Vendidos por Catagoria")
plt.xlabel("Total")
plt.ylabel("Categoria Do Produto");
```

```python
df.groupby(df["Data Do Pedido"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Do Pedido"].dt.year)["Lucro"].sum()
```

```python
#Selecionando apenas as vendas de 2012
df_2012 = df[df["Data Do Pedido"].dt.year == 2012]
```

```python
df_2012.head()
```

```python
df_2012.groupby(df_2012["Data Do Pedido"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2012.groupby("Categoria Do Produto")["Lucro"].sum().plot.bar(title="Lucro x Categoria")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
